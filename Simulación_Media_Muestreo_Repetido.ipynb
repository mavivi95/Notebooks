 # Simulación Media Muestreo Repetido

## Experimento A

```r
x <- c()
par(mfrow=c(2, 3))

for (i in 1:6){
  dist.normal = replicate(50*i, mean(rnorm(50, 5, 1)))
  hist(dist.normal, main = paste('Distribución Normal con', 50*i, '\nprocesos de muestreo'), 
        xlab = '', ylab = 'Frecuencia', col='#59B3D5')
}
```
