# Ejercicio Lanzar una Moneda

## Definición 1.
Se define  *frecuencia relativa* sí para un evento $A$ se tiene
\begin{align*}
    f(A) = \frac{n(A)}{n}
\end{align*}

Donde $n(A)$ es el número de veces que ocurre el evento $A$, y $n$ las veces que se repite el experimento.

## Observación 1.

Resulta que para un $A$ fijo, la relación $f(A)$ no es constante, ya que su valor depende de $n(A)$. Sin embargo, se ha observado que cuando se repite un experimento $n$ veces con $n\to \infty$ la frecuencia relativa $f(A)$ se estabiliza alrededor de un valor especifico entre 0 y 1.

Esto permite generar una medida de probabilidad $P$ del evento $A$, cuando un evento se repite muchas veces su probabilidad se aproxima a:

\begin{align*}
    P(A) = \lim_{n\to \infty}\frac{n(A)}{n}
\end{align*}

## Experimento A

Para este experimento se lanzará una moneda número variable de veces, y se observara el valor de la frecuencia relativa. Defineremos el evento $A$ sacar cara.

```r
moneda <- c(1, 0) # Espacio muestral, cara = 1, sello = 0
cont <- 0
freq <- c()
```

```r
# La función sample simula el resultado de lanzar la moneda, los parámetros de esta función 
# son el espacio muestral, el número de salida, y el vector de probabilidades de cada estado
# El resultado será 1 o 0 simulando caer en cara o sello

# Este fragmento de código es un ejemplo de como funciona la función sample
sample(moneda, 1, p=c(0.5,0.5)) 
```

```r
# En un ciclo for, se contaran las veces que la moneda cae en cara (1)
# este valor se acumulará en el vector cont, al mismo tiempo se acumulará en freq
# el valor del contador divido entre las veces que se repite el lanzamiento

#El barrido de las veces que se repite el experimento se realizará desde 1 vez hasta 1000

for (i in 1:500){
  if (sample(moneda, 1, p=c(0.5,0.5)) == 1){
    cont <- cont + 1
    }
  freq <- c (freq, cont/i)
}
```

```r
#Se realizará una gráfica con el valor de la frecuencia relativa vs el número de veces de
#repetir el experimento. 

plot(freq, main='Lanzamiento de una moneda - Experimento A', ylab = 'Frecuencia (evento cara)',
       type = 'o', col = '#69ABF5', lwd=2, ylim=c(0,1.5), xlab='Número de lanzamientos')
grid()
```

## Experimento B

Este experimento realiza el lanzamiento de una moneda 1000 veces y registra sus frecuencias relativas en un gráfico. Luego repite el proceso completo tres veces.

```r
moneda <- c(1, 0)
cont <- 0
freq <- c()

for (j in 1:3){
  for (i in 1:1000){
  if (sample(moneda, 1, p=c(0.5,0.5)) == 1){
    cont <- cont + 1
    }
  freq <- c (freq, cont/i)
  }
  plot(freq, main=paste('Lazamiento de una moneda', '\nExperimento B -', j), ylab = 'Frecuencia (evento cara)',
       type = 'o', col = '#69ABF5', xlab = 'Número de lanzamientos', lwd=2)
  grid()
  cont <- 0
  freq <- c()
}
```
